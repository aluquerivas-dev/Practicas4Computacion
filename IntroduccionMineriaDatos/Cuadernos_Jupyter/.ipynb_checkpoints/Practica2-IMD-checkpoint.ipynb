###  Pregunta [1,2]: Obtenga tres ejemplos de ficheros de datos en formato CSV, ARFF y usando Pandas cargue los ficheros y evalúe qué información puede obtener del histograma de atributos.

```python
from click._compat import raw_input
from scipy.io import arff
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

def normalizar(dataframe):
    dataframe = dataframe.iloc[:, :-1]
    normalized_df = (dataframe - dataframe.min()) / (dataframe.max() - dataframe.min())
    return normalized_df

def estandarizar(dataframe,μ,σ):
    dataframe = dataframe.iloc[:, :-1]
    standarized_df = (dataframe - μ) / (σ)
    return standarized_df

IRIS = arff.loadarff('./data/p2/iris.arff')
DIABETES = arff.loadarff('./data/p2/diabetes.arff')
GLASS = arff.loadarff('./data/p2/glass.arff')
df1 = pd.DataFrame(IRIS[0])
df2 = pd.DataFrame(DIABETES[0])
df3 = pd.DataFrame(GLASS[0])
#Pntamos IRIS
df1.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset iris.arff')

#Pintamos IRIS Normalizado
iris_normalizado = normalizar(df1)
iris_normalizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset iris.arff normalizado')

#Pintamos IRIS Estandarizado
iris_estandarizado = estandarizar(df1,df1.mean(),df1.std(axis=0))
iris_estandarizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset iris.arff estandarizado')

#Pintamos Matrix Iris
scatter_matrix(df1, alpha = 0.2, figsize = (4, 4),diagonal='kde')

#Correlacion Iris
plt.matshow(df1.corr())
plt.xticks(range(len(df1.columns)), df1.columns)
plt.yticks(range(len(df1.columns)), df1.columns)
plt.colorbar()

#Pintamos Matrix Normalizada
scatter_matrix(iris_normalizado, alpha = 0.2, figsize = (4, 4),diagonal='kde')
#Pintamos Matrix Estandarizada
scatter_matrix(iris_estandarizado, alpha = 0.2, figsize = (4, 4),diagonal='kde')

#Pintamos DIABETES
df2.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset diabetes.arff')

#Pintamos DIABETES Normalizado
diabetes_normalizado = normalizar(df2)
diabetes_normalizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset diabetes.arff normalizado')

#Pintamos DIABETES Estandarizado
diabetes_estandarizado = estandarizar(df2,df2.mean(),df2.std(axis=0))
diabetes_estandarizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset diabetes.arff estandarizado')

#Pintamos Matrix Diabetes
scatter_matrix(df2, alpha = 0.2, figsize = (8, 8),diagonal='kde')

#Correlacion Diabetes
plt.matshow(df2.corr())
plt.xticks(range(len(df2.columns)), df2.columns)
plt.yticks(range(len(df2.columns)), df2.columns)
plt.colorbar()

#Pintamos GLASS
df3.plot.hist(bins=12, alpha=0.5,by=4,label='Short label')
plt.suptitle('Histograma de dataset glass.arff')

#Pintamos GLASS Normalizado
glass_normalizado = normalizar(df3)
glass_normalizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset glass.arff normalizado')

#Pintamos GLASS Estandarizado
glass_estandarizado = estandarizar(df3,df3.mean(),df3.std(axis=0))
glass_estandarizado.plot.hist(bins=12, alpha=0.5,by=4)
plt.suptitle('Histograma de dataset glass.arff estandarizado')

#Pintamos Matrix GLASS
scatter_matrix(df3, alpha = 0.2, figsize = (9, 9),diagonal='kde')

#Correlacion GLASS
plt.matshow(df3.corr())
plt.xticks(range(len(df3.columns)), df3.columns)
plt.yticks(range(len(df3.columns)), df3.columns)
plt.colorbar()

plt.show()
```

### Pregunta [9]: Estudie la representación en coordenadas paralelas de los tres conjuntos e indique qué información relativa a las diferentes clases puede obtener.

```python
from click._compat import raw_input
from scipy.io import arff
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

IRIS = arff.loadarff('./data/p2/iris.arff')
DIABETES = arff.loadarff('./data/p2/diabetes.arff')
GLASS = arff.loadarff('./data/p2/glass.arff')
df1 = pd.DataFrame(IRIS[0])
df2 = pd.DataFrame(DIABETES[0])
df3 = pd.DataFrame(GLASS[0])

pd.plotting.parallel_coordinates(df1,'class')
plt.show()
pd.plotting.parallel_coordinates(df2,'class')
plt.show()
pd.plotting.parallel_coordinates(df3,'class')
plt.show()

```

### Respuesta [9]:

Las coordenadas paralelas son un método para explorar la propagación de
datos multidimensionales en una respuesta categórica, y echar un vistazo
a si hay alguna tendencia en las características. Podemos observar que claramente se muestran las propiedades para identificar  las clases de pertenencia en los gráficos de arriba.
